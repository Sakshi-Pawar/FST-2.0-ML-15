```python
#Python program to display the sum of n numbers using the list

n = int(input("Enter the number of elements: "))

l1 = []

for i in range(n):
    num = int(input("Enter a number: "))
    l1.append(num)
    
total = sum(l1)

print("The sum of the numbers:", total)
```

```python
#Python function which counts vowels and consonant in a word

str=input("Please enter string: ");
vowels=0
consonants=0
for i in str:
    if(i == 'a'or i == 'e'or i == 'i'or i == 'o'or i == 'u' or
       i == 'A'or i == 'E'or i == 'I'or i == 'O'or i == 'U' ):
           vowels=vowels+1;#vowel counter is incremented by 1
    else:
        consonants=consonants+1;
        
print("The number of vowels:",vowels);
print("\nThe number of consonant:",consonants);
```

```python
#b.	Which gender has better fitness levels?
#---> male gender has better fitness levels
avg_fit_by_gender = df.groupby("Gender")["Fitness"].mean()

# Print the result
print("Average fitness level by gender:",avg_fit_by_gender)
#print(average_fitness_by_gender)


import matplotlib.pyplot as plt

fit = df.groupby('Gender')['Fitness'].mean()

fig, ax = plt.subplots()
ax.bar(fit.index, fit.values)
ax.set_title('Fitness Level by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Fitness Level')

plt.show()
```

```python
#c.	Which is the most popular product among all?

count = df['Product'].value_counts()
most_popular_product = count.idxmax()
print("most popular product of the customers :", most_popular_product)
```

```python
#d.	Who is more fit? Married or singles
#---> as observed in bar graph Singles are more fit

import matplotlib.pyplot as plt

fit = df.groupby('MaritalStatus')['Fitness'].mean()

fig, ax = plt.subplots()
ax.bar(fit.index, fit.values)
ax.set_title('Fitness Level by Marital Status')
ax.set_xlabel('Marital Status')
ax.set_ylabel('Fitness Level')

plt.show()
```

```python
#e.	Which is the most favourite product of the customers below the age of 40?

df = df[df['Age'] < 40]
count = df['Product'].value_counts()
most_fav_product = count.idxmax()
print("most favourite product of the customers below the age of 40:", most_fav_product)
```
